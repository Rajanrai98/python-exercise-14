```python
def remove_duplicates(lst1,lst2):
    lst_new = []
    for items in lst1:
        if items in lst2:
            lst_new.append(items)
    return lst_new
```

```python
lst1 = [2,3,4,523,6,75,7,12]
lst2 = [2,4,7,9,7,12,9,23]
print(remove_duplicates(lst1,lst2))
```

```python
def gcd(x,y):
    smaller = min(x,y)
    gcd = 1
    for i in range(1,smaller+1):
        if x%i==0 and y%i==0:
            gcd = i
    return gcd

x = int(input("enter the value of x:"))
y = int(input("enter the value of y:"))
print(gcd(x,y))
              
```

```python
x = 36
y = 48 
lcm = (x*y)/(gcd(x,y))
print(lcm)
```

```python
import itertools
```

```python
def permutations(string):
    character = list(string)
    permutations = list(itertools.permutations(character))
    permutations = [" ".join(permutation) for permutation in permutations]
    
    return permutations
    
```

```python
string = "abc"
print(permutations(string))
```
